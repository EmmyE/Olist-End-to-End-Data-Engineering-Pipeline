Joining olist_marketing_qualified_leads_dataset.csv and olist_closed_deals_dataset.csv using mql_id

```python
from pyspark.sql.functions import col, when

# Load Your Datasets

leads_df = spark.read.csv(
    "abfss://demo@olist.dfs.core.windows.net/dbo.olist_marketing_qualified_leads_dataset.csv", 
    header=True, 
    inferSchema=True
)

closed_deals_df = spark.read.csv(
    "abfss://demo@olist.dfs.core.windows.net/dbo.olist_closed_deals_dataset.csv", 
    header=True, 
    inferSchema=True
)

# Perform the Left Join 
# We join 'all leads' (left) with 'closed deals' (right)
# The key for both is 'mql_id'
merged_df = leads_df.join(
    closed_deals_df, 
    on="mql_id", 
    how="left"
)

# Create a "Status" Column 
# If 'won_date' is NOT null, it means a match was found and the deal was closed.
# If 'won_date' IS null, it means the lead exists but was never closed.
analysis_df = merged_df.withColumn("lead_status", 
    when(col("won_date").isNotNull(), "Closed-Won")
    .otherwise("Open")
)

# Count the total for each status
final_counts_df = analysis_df.groupBy("lead_status").count()

print("Lead Status Breakdown:")
final_counts_df.show()
```

Visualizing the Lead Status with a Pie Chart

```python
import plotly.express as px

# Convert your small Spark DataFrame to Pandas
pandas_df = final_counts_df.toPandas()

# Create the interactive pie chart
fig = px.pie(
    pandas_df, 
    values='count',           # The column with your numbers
    names='lead_status',      # The column with your labels
    title='Lead Status: Closed-Won vs. Open',
    hover_data={'count': ':,d'} # formats the hover number
)

# Show the chart
fig.show()
```

Joining the 'closed_deals' and 'sellers' datasets will allow us to add demographic information to our leads and joining the result with 'order_items' will allow us to add monetary value.

```python
from pyspark.sql.functions import col, sum, count, first, desc

# Load All Three Datasets 

closed_deals_df = spark.read.csv(
    "abfss://demo@olist.dfs.core.windows.net/dbo.olist_closed_deals_dataset.csv", 
    header=True, 
    inferSchema=True
)

sellers_df = spark.read.csv(
    "abfss://demo@olist.dfs.core.windows.net/dbo.olist_sellers_dataset.csv", 
    header=True, 
    inferSchema=True
)

order_items_df = spark.read.csv(
    "abfss://demo@olist.dfs.core.windows.net/dbo.olist_order_items_dataset.csv", 
    header=True, 
    inferSchema=True
)

# Perform the Joins 

# Step 1: Join Closed Deals with Sellers (to get demographics)
# This adds seller_city, seller_state, etc. to your closed leads
deals_with_sellers = closed_deals_df.join(
    sellers_df,
    on="seller_id",
    how="left"
)

# Step 2: Join the result with Order Items (to get monetary value)
# This adds price, freight_value, order_id, etc.
# A single seller can have many orders, so this will create multiple rows
# for each seller, which is what we want for aggregation.
final_df = deals_with_sellers.join(
    order_items_df,
    on="seller_id",
    how="left"
)

# View the New Master Table 
print("Schema of the new joined table:")
final_df.printSchema()

print("Example data from the joined table:")
final_df.select("mql_id", "seller_id", "seller_city", "order_id", "price").show(10)
```

Now that we have the final_df, we can perform the monetary and demographic analysis.
We will now aggregate the data to make sense of it.
Monetary Analysis (Revenue per Closed Lead)
This code groups by each unique lead and sums up all their sales.


```python
# Group by the lead ID (mql_id) and the seller's city
revenue_per_lead = final_df.groupBy("mql_id", "seller_city").agg(
    sum("price").alias("total_revenue"),
    count("order_id").alias("total_orders")
)

print("Total revenue generated by each closed lead:")
revenue_per_lead.orderBy(col("total_revenue").desc()).show()
```

Demographic Analysis 
Let's see which cities the best leads are coming from.

```python
# Group by city to see where the most value is
value_by_city = revenue_per_lead.groupBy("seller_city").agg(
    sum("total_revenue").alias("city_total_revenue"),
    count("mql_id").alias("closed_leads_in_city")
)

print("Total value by seller city:")
value_by_city.orderBy(col("city_total_revenue").desc()).show()
```

We now need to generate a single, powerful table optimized for Power BI.
This approach combines all four tables into one denormalized "Fact Table." The grain (level of detail) of this table is one row per order_item. Leads who haven't closed, or closed deals that haven't sold anything, will have null values in the sales columns (like price). This is perfect, as it allows us to analyze your full lead funnel.


```python
from pyspark.sql.functions import col, when, datediff

# Load All Four Datasets 

leads_df = spark.read.csv("abfss://demo@olist.dfs.core.windows.net/dbo.olist_marketing_qualified_leads_dataset.csv", header=True, inferSchema=True)
closed_deals_df = spark.read.csv("abfss://demo@olist.dfs.core.windows.net/dbo.olist_closed_deals_dataset.csv", header=True, inferSchema=True)
sellers_df = spark.read.csv("abfss://demo@olist.dfs.core.windows.net/dbo.olist_sellers_dataset.csv", header=True, inferSchema=True)
order_items_df = spark.read.csv("abfss://demo@olist.dfs.core.windows.net/dbo.olist_order_items_dataset.csv", header=True, inferSchema=True)

# Perform the Series of Left Joins 

# Start with ALL leads
all_leads_df = leads_df.join(
    closed_deals_df,
    on="mql_id",
    how="left"
)

# Add seller demographic data
leads_with_sellers_df = all_leads_df.join(
    sellers_df,
    on="seller_id",
    how="left"
)

# Add all order item data (this creates the final grain)
final_analysis_table = leads_with_sellers_df.join(
    order_items_df,
    on="seller_id",
    how="left"
)

# Add Calculated Columns for BI 

final_analysis_table = final_analysis_table.withColumn(
    # A clean "Status" dimension for filtering
    "lead_status",
    when(col("won_date").isNotNull(), "Closed-Won")
    .otherwise("Open")
)

final_analysis_table = final_analysis_table.withColumn(
    # A key KPI: Time to close a lead
    "time_to_close_days",
    datediff(col("won_date"), col("first_contact_date"))
)

# Select Only the Important Columns 

power_bi_table = final_analysis_table.select(
    # Lead Info
    "mql_id",
    "first_contact_date",
    "origin",
    "landing_page_id",
    
    # Conversion Info
    "lead_status",
    "won_date",
    "time_to_close_days",
    "business_segment",
    "lead_type",
    
    # Seller Info
    "seller_id",
    "seller_city",
    "seller_state",
    "seller_zip_code_prefix",
    
    # Sales Info (Measures)
    "order_id",
    "order_item_id",
    "product_id",
    "price",
    "freight_value"
)

# Save the Table for Power BI 
power_bi_table.write.mode("overwrite").parquet("abfss://demo@olist.dfs.core.windows.net/gold/PBI_Lead_Performance")

print("Power BI table created successfully.")
power_bi_table.printSchema()
```
